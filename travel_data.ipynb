```
import re
#文章の整形関数
def format_text(text):

    text=re.sub(r'https?://[\w/:%#\$&\?\(\)~\.=\+\-…]+', "", text)
    text=re.sub('あわせて読む', "", text)
    text=re.sub('データ提供', "", text)
    text=re.sub('出典', "", text)
    text=re.sub('さんの投稿', "", text)
    text=re.sub('円', "", text)
    ext=re.sub('〜', "", text)
    ext=re.sub('人', "", text)
    ext=re.sub('引用元', "", text)
    text=re.sub(r'[!-~]', "", text)#半角記号,数字,英字
    text=re.sub(r'[︰-＠]', "", text)#全角記号
    #text=re.sub('\n', " ", text)#改行文字

    return text
```
